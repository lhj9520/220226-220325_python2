```python
# 폰트 지정 - 한글 사용시
import matplotlib.font_manager as fm
import matplotlib

font_loacation = 'C:\Windows\Fonts\gulim.ttc'
fname = fm.FontProperties(fname=font_loacation).get_name()
matplotlib.rc('font',family = fname)
# 간격을 가능한 최적의 수치로 자동 조정해준다.
# fig = plt.figure(constrained_layout=True)
fig = plt.figure()
# 그래프 위치 지정
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
#원하는 위치에 그래프를 그림
x = range(10)
y = range(5,15)
ax1.plot(x,y,"b.")
ax2.plot(y,x,"r^")
# x축 y축 label
ax1.set_xlabel('개수')
ax1.set_ylabel('합계')
ax2.set_xlabel('count')
ax2.set_ylabel('sum')
plt.show()
```

```python
# 막대 그래프 생성
plt.style.use('ggplot')
fig = plt.figure()
#subplot(행 개수, 열 개수, 위치값)
ax = fig.add_subplot(1,1,1)
colors = ['red','green']
labels = ['a','b','c','d']
x = [2,4,6,8]
y = [1,4,9,5]
ax.bar(x,y, tick_label = labels, color=colors)
# ax.barh(x,y, tick_label = labels, color=colors) #막대 수평
for i,v in enumerate(y):
    ax.text(v, i, str(v), color='blue')
plt.show()
```

```python
import numpy as np

x = [1,2]
y1,y2,y3 = [1,6],[2,4],[3,8]
fig = plt.figure()
ax = fig.add_subplot(111)
w = 0.2
ax.bar(x,y1,width=w,label="a")
ax.bar(np.array(x)+w,y2,width=w,label="b")
ax.bar(np.array(x)+w*2,y3,width=w,label="c")
ax.legend()
```

```python
# 원차트(pie차트)
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
label = ["english","math","korean"]
sizes = [45,35,20]
ax.pie(sizes, labels = label)
plt.show()
```
