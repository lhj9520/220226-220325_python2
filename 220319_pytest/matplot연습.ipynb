```python
# pip install selenium
# pip install matplotlib

import matplotlib.pyplot as plt

# x축 좌표와 y축 좌표를 지정하지 않을 경우
# x축은 0부터 1씩 증가, y축은 사용자가 지정한 값으로 인식
plt.style.use('default')
plt.plot([2,4,5,9])
plt.show()
```

```python
#x축, y축 좌표 지정
x = [2,4,6,8]
y = [1,4,9,5]
plt.plot(x,y)
plt.show()
```

```python
# matplot option
# color 수정/그래프유형
# [색상표] r,g,b,m(마젠타),y,k(검정),w(흰색)
# [마커표] o 원, v 역삼각형, ^ 삼각형, s 사각형, + 플러스, . dot, default line
x = range(10)
y = range(5,15)
plt.plot(x,y,"k.")
plt.show()
```
